```python
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
print('Number of numerical features are :',len(numeric_features))
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('ProdTaken',axis = 1)
Y = df['ProdTaken']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 42)
X_train.shape,X_test.shape
```

```python
##categorical feartures now need to be converted into numeric values
```

```python
categorical_features = X.select_dtypes(include = "object").columns
numeric_features = X.select_dtypes(exclude = "object").columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Assuming you have already defined your transformers and feature lists
oh_transformer = OneHotEncoder()
numeric_transformer = StandardScaler()

preprocessor = ColumnTransformer([
    ('OneHotEncoder', oh_transformer, categorical_features),
    ('StandardScaling', numeric_transformer, numeric_features)
])

# Fit and transform the training data
X_train_transformed = preprocessor.fit_transform(X_train)

# Only transform the test data (no fitting)
X_test_transformed = preprocessor.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score,recall_score ,confusion_matrix,accuracy_score,classification_report ,f1_score ,roc_curve,roc_auc_score
```

```python
models = {
    "Random Forest": RandomForestClassifier()
}
```

```python
for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train_transformed,Y_train)
```

```python
y_train_pred = model.predict(X_train_transformed)
y_test_pred = model.predict(X_test_transformed)
```

```python
model_train_accuracy = accuracy_score(Y_train,y_train_pred)
model_test_accuracy = accuracy_score(Y_test,y_test_pred)

model_train_f1 = f1_score(Y_train,y_train_pred)
model_test_f1 = f1_score(Y_test,y_test_pred)

model_train_precision = precision_score(Y_train,y_train_pred)
model_test_precision = precision_score(Y_test,y_test_pred)

model_train_recall = recall_score(Y_train,y_train_pred)
model_test_recall = recall_score(Y_test,y_test_pred)

model_train_roc = roc_auc_score(Y_train,y_train_pred)
model_test_roc = roc_auc_score(Y_test,y_test_pred)
```
